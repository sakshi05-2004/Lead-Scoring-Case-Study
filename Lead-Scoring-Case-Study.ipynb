__Columns with Very High Missing Values (Above 40%)__

* These columns have too many missing values, and keeping them may introduce bias. We will consider dropping them.

__Columns with Moderate Missing Values (20%-40%)__

* For categorical columns, we can fill missing values with "Unknown" or "Other" to retain information.

__Columns with Low Missing Values (Below 20%)__

* For numerical columns, we can fill missing values with the median.
* For categorical columns, we can fill with the most frequent value (mode).

<h2>Data Cleaning<h2></h2>

```python
# Dropping columns with more than 40% missing values
df = df.drop(columns=['Lead Quality', 'Asymmetrique Profile Score', 'Asymmetrique Activity Score', 
                      'Asymmetrique Profile Index', 'Asymmetrique Activity Index'])
```

```python
# Filling categorical missing values with 'Unknown'
fill_unknown_cols = ['Tags', 'Lead Profile', 'What matters most to you in choosing a course', 
                     'What is your current occupation', 'Country', 'How did you hear about X Education']
df.loc[:, fill_unknown_cols] = df[fill_unknown_cols].fillna('Unknown')
```

```python
# Filling missing values in categorical columns with mode (most frequent value)
fill_mode_cols = ['Specialization', 'City', 'Last Activity', 'Lead Source']
df.loc[:, fill_mode_cols] = df[fill_mode_cols].fillna(df[fill_mode_cols].mode().iloc[0])
```

```python
# Filling missing numerical values with median
fill_median_cols = ['TotalVisits', 'Page Views Per Visit']
df.loc[:, fill_median_cols] = df[fill_median_cols].fillna(df[fill_median_cols].median())
```

```python
# Verifying if any missing values remain
print("Missing values after handling:", df.isnull().sum().sum())
```

__Handling Missing Values in Categorical Columns__

```python
missing_categorical_cols = [col for col in categorical_cols if df[col].isnull().sum() > 0]

print("Categorical columns with missing values and their percentage:")
print(df[missing_categorical_cols].isnull().mean() * 100)
```

```python
# Fill missing values
for col in missing_categorical_cols:
    if df[col].nunique() > 5:  # More categories, fill with 'Unknown'
        df[col].fillna('Unknown', inplace=True)
    else:  # Fewer categories, fill with mode (most frequent value)
        df[col].fillna(df[col].mode()[0], inplace=True)

# Verify if missing values are handled
print("\nMissing values after handling:")
print(df[missing_categorical_cols].isnull().sum().sum())
```

__Encoding Categorical Variables (Dummy Variables)__

```python
from sklearn.preprocessing import LabelEncoder

# Drop unique identifier columns
df.drop(['Prospect ID', 'Lead Number'], axis=1, inplace=True)
```

```python
# Identify binary categorical columns (with only 2 unique values)
binary_cols = [col for col in categorical_cols if df[col].nunique() == 2]

# Convert binary categorical variables into 0 and 1
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
# Identify multi-category columns for one-hot encoding
multi_category_cols = [col for col in categorical_cols if df[col].nunique() > 2]

# Apply one-hot encoding (convert categories into separate columns)
df = pd.get_dummies(df, columns=multi_category_cols, drop_first=True)

# Print the new shape of the dataset
print("Shape of dataset after encoding:", df.shape)
```

__Feature Selection (Removing High Correlation)__

```python
# Check for non-numeric columns
non_numeric_cols = df.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols.tolist())
```

```python
# Drop columns that contain only one unique value
drop_cols = ['Magazine', 'Receive More Updates About Our Courses', 
             'Update me on Supply Chain Content', 'Get updates on DM Content', 
             'I agree to pay the amount through cheque']

df.drop(columns=drop_cols, inplace=True)

print("Dropped non-informative categorical columns.")
```

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Identify and remove highly correlated features (threshold: 0.9)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_correlation_features = [column for column in upper.columns if any(upper[column] > 0.9)]

# Drop highly correlated features
df.drop(columns=high_correlation_features, inplace=True)

print("Shape after removing highly correlated features:", df.shape)
```

```python
from sklearn.model_selection import train_test_split

# Define target variable
y = df['Converted']

# Define feature variables (drop target)
X = df.drop(columns=['Converted'])
```

```python
# Split into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Print dataset shapes
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

<h2>Feature Scaling<h2></h2>

__Standardizing the Data__

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Apply scaling only on numerical features
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Identify numerical columns
num_cols = X_train.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Fit and transform training data, transform test data
X_train_scaled[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test_scaled[num_cols] = scaler.transform(X_test[num_cols])

# Print summary after scaling
print("Feature scaling complete!")
```

<h2>Model Building (Logistic Regression)<h2></h2>

__Training the Model__

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

# Initialize Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on training data
log_reg.fit(X_train_scaled, y_train)
```

```python
# Make predictions
y_train_pred = log_reg.predict(X_train_scaled)
y_test_pred = log_reg.predict(X_test_scaled)

# Evaluate performance
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
roc_auc = roc_auc_score(y_test, log_reg.predict_proba(X_test_scaled)[:, 1])

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
print(f"ROC-AUC Score: {roc_auc:.4f}")
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```
